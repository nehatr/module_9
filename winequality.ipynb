In the exploratory analysis below, I explore the wine quality dataset using the plotly and plotly express libraries. All referenced code precedes the modified code.

```python
df = pd.read_csv("winequality-red.csv",sep=";")
df.describe()
```

```python

# The pandas corr() method returns a dataframe describing correlations between each feature.
z= df.corr()

# Convert the dataframe to a matrix. In the next step, the image show method (from plotly express) is used to 
z =z.as_matrix()


fig = px.imshow(z,labels=dict(color="Correlation"),
                x=['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
               'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
               'pH', 'sulphates', 'alcohol', 'quality'],
                
               y = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
               'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
               'pH', 'sulphates', 'alcohol', 'quality'],
                
                color_continuous_scale=px.colors.sequential.Plasma)  ## Modified

fig.update_layout(title_text= "Correlation Matrix: Wine Quality",title_x=0.5)   ## Modified code
fig.show()
```

Modifications made to code:
- referenced dataframe columns for the values of x, y, z and labels 
- specified the color scale as plasma
- added title text
- specified the height and width of the graph  
- changed the alignment of the graph on the page


```python
x, y, z = df["alcohol"], df['sulphates'], df['citric acid']
labels =  df['quality']
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=8,
        color=labels,             
        colorscale='plasma',   
        opacity=0.8
    )
)])

fig.update_layout(
    margin=dict(l=50, r=50, b=50, t=50),
    title_text="Wine Quality : ",
    width = 800,
    height = 600)

fig.show()
```
